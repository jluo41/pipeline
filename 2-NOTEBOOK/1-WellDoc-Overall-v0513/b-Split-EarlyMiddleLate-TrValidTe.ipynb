```python
# # ---- STEP 6: Within-patient Temporal Splits ----
# df['ObsDT'] = pd.to_datetime(df['ObsDT'])
# split_date_dict = dict()


def label_time_bin(df_one_patday, 
                   pct=0.2, 
                   middle_range = [1,3], 
                   late_range = [1,20]):
    
    # group = group.sort_values("ObsDT")
    df_one_patday = df_one_patday.sort_values('Date').reset_index(drop = True)
    n = len(df_one_patday)

    # Late only between 3-10 days
    # Late at most 20 days. 
    # Middle only 1 days. 
    if n <= 15:
        d = {}
        d['PID'] = df_one_patday.iloc[0]['PID']
        if n >= 5:
            d['middle_first_date'] = df_one_patday.iloc[-2]['Date'] + pd.Timedelta(days=1) # 
            d['late_first_date'] = df_one_patday.iloc[-1]['Date'] + pd.Timedelta(days=1)
            d['n_early_days'] = n - 1
            d['n_middle_days'] = 1
            d['n_late_days'] = 0
        else:
            d['middle_first_date'] = df_one_patday.iloc[-1]['Date'] + pd.Timedelta(days=1)
            d['late_first_date'] = df_one_patday.iloc[-1]['Date'] + pd.Timedelta(days=1)
            d['n_early_days'] = n
            d['n_middle_days'] = 0
            d['n_late_days'] = 0
    
    else:
        #split_idx = int((1 - pct) * n) # split_idx cut off point for middle and late. 
        n_late = int((pct) * n)
        # decide number of late days:
        if n_late>late_range[1]: # sufficient late days
            n_late = late_range[1]
            n_middle = middle_range[1]
        elif n_late<3: # not enough late days
            n_late = late_range[0]
            n_middle = 0
        else:
            # moderate late days
            n_middle = int((middle_range[1]+middle_range[0])/2)
        
        # n_early, n_middle, n_late # n is the good day. 
        n_early = n - n_middle - n_late 
        
        assert n_early>10 and n_early+n_middle+n_late==n
        # early (>10 days), middle (<=3 day or none, get from early), late (1-20 days or none)
        # group['time_bin'] = ['early'] * n_early + ['middle']*n_middle + ['late'] * n_late
        
        # Get the split date: first date in "late"
        split_idx = n_early+n_middle+1
        # late_start_date_index = split_idx
        
        d = {}
        d['PID'] = df_one_patday.iloc[0]['PID']
        d['middle_first_date'] = df_one_patday.iloc[n_early+1]['Date'] + pd.Timedelta(days=1)
        d['late_first_date'] = df_one_patday.iloc[split_idx]['Date'] + pd.Timedelta(days=1)

        d['n_early_days'] = n_early
        d['n_middle_days'] = n_middle
        d['n_late_days'] = n_late

    return d

# df = df.groupby('PID', group_keys=False, observed=True).apply(label_time_bin)
# #store split data to df_patient
# df_patient['split_date'] = df_patient['PID'].map(split_date_dict)
```

```python

# Define a function to map age to age group
def map_age_to_group(age):
    """
    Maps an age value to an age group category.
    
    Args:
        age (int): The age to categorize
        
    Returns:
        str: Age group category ("0-17", "18-39", "40-64", or "65+")
    """
    if age < 18:
        return "0-17"
    elif age < 40:
        return "18-39"
    elif age < 65:
        return "40-64"
    else:
        return "65+"
    

tag_columns = [
    'PID', #'ObsDT', 
    'YearOfBirth', 'UserTimeZone', 
    'UserTimeZoneOffset', 
    'Gender', 'MRSegmentID', 'DiseaseType']
    
df_patient = df_case_all[tag_columns].drop_duplicates().reset_index(drop = True)
print(df_patient.shape)
df_patient 



# df_patient['age'] = (df_patient['Date'].dt.year - df_patient['YearOfBirth']).astype(int)
df_patient['age'] = 2024 - df_patient['YearOfBirth']
df_patient['age_group'] = df_patient['age'].apply(map_age_to_group)
df_patient
```

```python
df_case_day = df_case_hours[df_case_hours['ObsDT'].dt.hour == 0].reset_index(drop = True)

columns = ['PID', 'ObsDT', 
'MEDInfoBf24h-MedRecNum', 
'MEDInfoBf24h-ExerciseRecNum', 
'MEDInfoBf24h-DietRecNum',
]

df_case_day = df_case_day[columns]
df_case_day['Date'] = df_case_day['ObsDT'] - pd.Timedelta(days=1)
df_case_day.columns = ['PID', 'ObsDT', 'MedRecNum', 'ExerciseRecNum',
       'DietRecNum', 'Date']


df_case_day = df_case_day[['PID', 'Date', 'MedRecNum', 'ExerciseRecNum', 'DietRecNum']]
df_case_day['has_med'] = (df_case_day['MedRecNum'] > 0).astype(int)
df_case_day['has_exercise'] = (df_case_day['ExerciseRecNum'] > 0).astype(int)
df_case_day['has_diet'] = (df_case_day['DietRecNum'] > 0).astype(int)
df_case_day['has_event'] = (df_case_day[['MedRecNum', 'ExerciseRecNum', 'DietRecNum']].sum(axis=1) > 0).astype(int)
df_case_day
# Patient Date

df_pat_event = df_case_day.groupby('PID').agg(
    {   
        'MedRecNum': 'sum',
        'ExerciseRecNum': 'sum',
        'DietRecNum': 'sum',
        'has_med': 'sum',
        'has_exercise': 'sum',
        'has_diet': 'sum',
        'has_event': 'sum',
    }
).reset_index()

df_pat_event
```

```python
df_patient_info["stratum"] = (
    df_patient_info[["age_group", "Gender", "DiseaseType"]]
    .astype(str)
    .agg("_".join, axis=1)
)

df_patient_info['has_good_days_geq_10'] = (df_patient_info['n_good_days'] >= 10).astype(str)
df_patient_info["event_category"] = (df_patient_info["has_event"] > 0).astype(str) # pd.qcut(df_patient["event_ratio"], 3, labels=["low", "medium", "high"], duplicates="drop")
df_patient_info["event_stratum"] = df_patient_info["stratum"] + "_" + "event."+ df_patient_info["event_category"].astype(str)

df_patient_info["event_good_days_stratum"] = df_patient_info["stratum"] + "_" + "event."+ df_patient_info["event_category"].astype(str) + "_10gooddays" + df_patient_info["has_good_days_geq_10"].astype(str)

df_patient_info["good_days_stratum"] = df_patient_info["stratum"] +  "_10gooddays" + df_patient_info["has_good_days_geq_10"].astype(str)

df_patient_info
```

```python
# ---- STEP 4: Filter valid strata ----
stratum_counts = df_patient_info["stratum"].value_counts()
selected_strata = stratum_counts[stratum_counts >= 20].index
df_patient_filtered = df_patient_info[df_patient_info["stratum"].isin(selected_strata)]
print(len(df_patient_filtered), 'Kept out of total', len(df_patient_info))
selected_strata
```

```python
# df_patient_filtered[]
```

```python
# ---- STEP 5: Hold-in / Hold-out Split at Patient Level ----
# Use combined stratum with event category for better stratification
from sklearn.model_selection import train_test_split

stratum_column = 'good_days_stratum'
test_pct = 0.15
val_pct  = 0.1
random_seed = 42
# First: Hold out the test set
hold_in_pids, test_pids = train_test_split(
    df_patient_filtered['PID'],
    test_size=test_pct,
    stratify=df_patient_filtered[[stratum_column]],
    random_state=random_seed
)

# Subset for hold-in patients
df_hold_in = df_patient_filtered[df_patient_filtered['PID'].isin(hold_in_pids)]

# Second: Split hold-in into train and val
train_pids, val_pids = train_test_split(
    df_hold_in['PID'],
    test_size=val_pct / (1 - test_pct),  # rescale val proportion
    stratify=df_hold_in[[stratum_column]],
    random_state=random_seed
)

df_patient_info['split'] = 'unassigned'
df_patient_info.loc[df_patient_info['PID'].isin(train_pids), 'split'] = 'train'
df_patient_info.loc[df_patient_info['PID'].isin(val_pids), 'split'] = 'val'
df_patient_info.loc[df_patient_info['PID'].isin(test_pids), 'split'] = 'test'


# Final sanity check
assert len(train_pids) + len(val_pids) + len(test_pids) == len(df_patient_filtered)
print(len(train_pids), len(val_pids), len(test_pids), '---', len(df_patient_filtered))
```
